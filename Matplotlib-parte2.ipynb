# Gráfico de Dispersão

## Importando Dados e pacotes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

ax.scatter(df['comprimento_sépala'], df['largura_sépala'])
ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    ax.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                 color=cores[especie])
ax.legend(cores, fontsize=20)
ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)
ax.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'r', 'Iris-versicolor': 'b', 'Iris-virginica': 'g'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    ax.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
                 color=cores[especie], marker=marcadores[especie],
                 s=100)
ax.legend(cores, fontsize=20)
ax.set_title('Gráfico de dispersão', fontsize=25, pad=15)
ax.set_xlabel('Comprimento da sépala', fontsize=15)
ax.set_ylabel('Largura da sépala', fontsize=15)
ax.tick_params(labelsize=15)
```

# BoxPlot

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.boxplot(df['largura_pétala'].values, patch_artist=True)
ax.set_title('Gráfico de caixa', fontsize=15, pad=10)
ax.set_xticklabels(['largura_pétala'])
```

```python

fig = plt.figure(figsize=(8,5))
ax = fig.add_axes([0,0,1,1])

ax.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
ax.set_title('Gráfico de caixa', fontsize=15, pad=10)
ax.set_xticklabels(df.drop('espécie', axis=1).columns)
ax
```

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_axes([0,0,1,1])

cores = ['red', 'blue', 'orange', 'green']

caixas = ax.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)
ax.set_title('Gráfico de caixa', fontsize=15, pad=10)
ax.set_xticklabels(df.drop('espécie', axis=1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color=cor)

for outlier in caixas['fliers']:
    outlier.set(marker='x', markersize=8)
```

# Gráfico Histograma

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.hist(df['comprimento_pétala'], color= 'orange')
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.hist(df['comprimento_pétala'], bins=20, orientation='horizontal')
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.hist(df['comprimento_pétala'], bins=20, density=True, color= 'orange')
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)
```

```python

fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

ax.hist(df['comprimento_pétala'], bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.5, 20), fontsize=20)

ax.axvline(mu, color = 'k', linestyle='--')
ax.annotate('média', xy=(mu-1.3, 28), fontsize=20)

ax.axvline(df['comprimento_pétala'].median(), color='g', linestyle='--')
ax.annotate('mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=20)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

df_is = df[df['espécie'] == 'Iris-setosa']

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

ax.hist(df_is['comprimento_pétala'], bins=20)
ax.set_title('Iris Setosa', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(1.7, 7), fontsize=20)

ax.axvline(mu, color = 'k', linestyle='--')
ax.annotate('média', xy=(mu-0.2, 13.5), fontsize=20)

ax.axvline(df_is['comprimento_pétala'].median(), color='g', linestyle='--')
ax.annotate('mediana', xy=(df_is['comprimento_pétala'].median()+0.05, 12), fontsize=20,
               color='g')

fig.savefig('histograma_is.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

df_ivc = df[df['espécie'] == 'Iris-virginica']

mu, sigma = df_ivc['comprimento_pétala'].mean(), df_ivc['comprimento_pétala'].std()

ax.hist(df_ivc['comprimento_pétala'], bins=20)
ax.set_title('Iris Virginca', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(6.3, 7), fontsize=20)

ax.axvline(mu, color = 'k', linestyle='--')
ax.annotate('média', xy=(mu, 8), fontsize=20)

ax.axvline(df_ivc['comprimento_pétala'].median(), color='g', linestyle='--')
ax.annotate('mediana', xy=(df_ivc['comprimento_pétala'].median(), 7), fontsize=20,
               color='g')

fig.savefig('histograma_iv.png', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

ax.hist(df_iv['comprimento_pétala'], bins=20)
ax.set_title('Iris Versicolor', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
               xy=(4.6, 6), fontsize=20)

ax.axvline(mu, color = 'k', linestyle='--')
ax.annotate('média', xy=(mu-0.5, 5.5), fontsize=20)

ax.axvline(df_iv['comprimento_pétala'].median(), color='g', linestyle='--')
ax.annotate('mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=20,
               color='g')

fig.savefig('histograma_ivg.png', bbox_inches='tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new("RGB", (3*largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png',
                               'histograma_ivg.png']):
    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura

combinada.save('combinada.png')
```
