# Visualização com Matplotlib

## Importando os dados e explorando os dados

```python
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Convertendo a coluna data para o tipo desejado

```python
df['data'] = pd.to_datetime(df['data'])
df.info()
```

## Vamos plotar em um gráfico a variação de temperatura ao longo do tempo.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(df['data'], df['temperatura'], color= 'orangered')

ax.set_title('Temperatura em 2014', fontsize=25)
ax.set_ylabel('Temperatura', fontsize=20)
ax.set_xlabel('Data', fontsize=20)
ax.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

## importando o datetime

```python
import datetime

fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(df['data'], df['temperatura'], color = 'r')

ax.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
ax.set_title('Temperatura no momento', fontsize=25)
ax.set_ylabel('Temperatura', fontsize=20)
ax.set_xlabel('Data', fontsize=20)
ax.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))

ax = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.7, 0.65, 0.3, 0.3])

ax.grid(True)
ax.plot(df['data'], df['temperatura'], color = 'g')
ax.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
ax.set_title('Temperatura em 2014', fontsize=25, pad=20)
ax.legend(['temperatura'], loc='lower right', fontsize=15)
ax.set_ylabel('Temperatura', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

ax2.plot(df['data'], df['temperatura'], color = 'b')
ax2.set_title('Temperatura no momento', fontsize=15)
ax2.legend(['temperatura'], loc='best', fontsize=8)
ax2.set_ylabel('Temperatura', fontsize=10)
ax2.set_xlabel('Data', fontsize=10)
ax2.grid(True)
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.7,0.65,0.3,0.3])

ax.grid(True)
ax.plot(df['data'], df['temperatura'], color = 'g')
ax.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
ax.set_title('Temperatura em 2014', fontsize=25)
ax.legend(['Temperatura'], loc = 'lower right', fontsize=15)
ax.set_ylabel('Temperatura', fontsize=20)
ax.set_xlabel('Data', fontsize=20)

ax2.grid(True)
ax2.plot(df['data'], df['temperatura'], color = 'b')
ax2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
ax2.set_title('Temperatura no momento', fontsize=15)
ax2.legend(['Temperatura'], loc = 'best', fontsize=8)
ax2.set_ylabel('Temperatura', fontsize=10)
ax2.set_xlabel('Data', fontsize=10)
```

## Limitando o ponto máximo e o ponto mínimo utilizando barras.

```python
fig = plt.figure(figsize = (15,8))
ax = fig.add_axes([0,0,1,1])
ax.plot(df['data'], df['temperatura'], color = 'b')
ax.set_title('Temperatura no momento', fontsize = 25, pad = 20)
ax.set_xlabel('Data', fontsize=20)
ax.set_ylabel('Temperatura', fontsize=20)
ax.legend(['Temperatura'], loc='lower right', fontsize=15)
ax.grid(True)
ax.axhline(max(df['temperatura']), color = 'r', linestyle='--')
ax.axhline(min(df['temperatura']), color = 'r', linestyle='--')
```

# Gráfico de Barras

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
temperatura_por_dia_da_semana 
```

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_da_semana))

ax.bar(indice, temperatura_por_dia_da_semana)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
ax.set_xlabel('Dia da semana', fontsize=15)
ax.set_ylabel('Temperatura média', fontsize=15)
ax.set_xticks(indice)
ax.set_xticklabels(nome_dias)
ax
```

# Gráfico de Pizza

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%', explode=explodir)
ax.set_title('Temperatura por dia da semana', size=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%', explode=explodir, shadow=True)
ax.set_title('Temperatura por dia da semana', size=15, pad=10)
```
